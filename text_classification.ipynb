```python
category_mapping = {
'Fiction' : "Fiction",
 'Juvenile Fiction': "Children's Fiction",
 'Biography & Autobiography': "Nonfiction",
 'History': "Nonfiction",
 'Literary Criticism': "Nonfiction",
 'Philosophy': "Nonfiction",
 'Religion': "Nonfiction",
 'Comics & Graphic Novels': "Fiction",
 'Drama': "Fiction",
 'Juvenile Nonfiction': "Children's Nonfiction",
 'Science': "Nonfiction",
 'Poetry': "Fiction"
 }

books['simple_categories'] = books['categories'].map(category_mapping)
```

```python
books
```

```python
books[books['simple_categories'].isna()]
```

```python
from transformers import pipeline

categories = ['Fiction', 'Nonfiction']
classifier = pipeline('zero-shot-classification', 
                       model="facebook/bart-large-mnli")
```

```python
sequence_fiction = books.loc[books['simple_categories'] == 'Fiction', 'description'].reset_index(drop=True)
sequence_nonfiction = books.loc[books['simple_categories'] == 'Nonfiction', 'description'].reset_index(drop=True)
```

```python
sequence = sequence_fiction[0]
response = classifier(sequence, categories)
```

```python
index = np.argmax(response['scores'])
result = response['labels'][index]
result
```

```python
def classify_text(text, categories):
    response = classifier(text, categories)
    index = np.argmax(response['scores'])
    result = response['labels'][index]
    return result
```

```python
from tqdm import tqdm

actual_category = []
predicted_category = []

for i in tqdm(range(0, 300)):
    sequence = sequence_fiction[i]
    predicted_category += [classify_text(sequence, categories)]
    actual_category += ['Fiction']

for i in tqdm(range(0, 300)):
    sequence = sequence_nonfiction[i]
    predicted_category += [classify_text(sequence, categories)]
    actual_category += ['Nonfiction']

```

```python
predictions_df = pd.DataFrame({'actual_category': actual_category, 'predicted_category': predicted_category})
```

```python
predictions_df
```

```python
predictions_df['correct_predictions'] = np.where(predictions_df['actual_category'] == predictions_df['predicted_category'], 1, 0)
```

```python
predictions_df['correct_predictions'].sum()/len(predictions_df)
```

```python
isbns = []
predicted_cats = []

missing_cats = books.loc[books['simple_categories'].isna(), ['isbn13', 'description']].reset_index(drop=True)
```

```python
for i in tqdm(range(0, len(missing_cats))):
    sequence = missing_cats['description'][i]
    predicted_cats += [classify_text(sequence, categories)]
    isbns += [missing_cats['isbn13'][i]]
```

```python
missing_predicted_df = pd.DataFrame({'isbn13':isbns, 'predicted_categories':predicted_cats})
```

```python
missing_predicted_df
```

```python
books = pd.merge(books, missing_predicted_df, on='isbn13', how='left')
books['simple_categories'] = np.where(books['simple_categories'].isna(), books['predicted_categories'], books['simple_categories'])
```

```python
books = books.drop(columns = ["predicted_categories"])
```
